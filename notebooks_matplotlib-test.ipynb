```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(10)
```

```python
N=30
x=np.random.rand(N)
y=np.random.rand(N)
colors=np.random.rand(N)
```

```python
area=(30*np.random.rand(N))
plt.scatter(x,y,s=area,c=colors,alpha=0.4)
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
x=[2,4,6]
y=[12,14,16]
x2=[3,3,4]
y2=[7,14,5]
```

```python
plt.bar(x,y,color="r",align="center")
plt.bar(x2,y2,color="b",align="center")
plt.title('Info')
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```
